```python
# import libraries
import pandas as pd
import re
import numpy as np

pd.options.display.max_colwidth = 1000 # display full text in a cell
```

```python
# load data
data = pd.read_csv('egrz.csv', delimiter=';',low_memory=False)
```

```python
# rename colums
data = data.rename(columns={'Идентификатор':'id',
                           'Номер заключения экспертизы':'expert_id',
                           'Дата заключения экспертизы':'expert_date',
                           'Результат проведенной экспертизы (положительное или отрицательное заключение экспертизы)':'expert_result',
                           'Форма экспертизы (государственная, негосударственная)':'expert_type',
                           'Сведения об объекте экспертиз (проектная документация и(или) результаты инженерных изысканий)':'doc_type',
                           'Субъект РФ':'region',
                           'Код субъекта РФ':'region_id',
                           'Наименование и адрес (местоположение) объекта капитального строительства, применительно к которому подготовлена проектная документация':'object',
                           'Сведения об экспертной организации':'expert_name',
                           'Сведения о застройщике, обеспечившем подготовку проектной документации':'builder_name',
                           'Сведения о техническом заказчике, обеспечившем подготовку проектной документации':'tech_cust_name',
                           'Сведения об индивидуальных предпринимателях и (или) юридических лицах, подготовивших проектную документацию':'designer_name',
                           'Сведения об использовании типовой проектной документации':'typical',
                            'Сведения о типовой проектной документации':'typical2',
                           'Признак типовой проектной документации':'typ_id',
                           'Дата включения сведений в реестр':'reestr_date',
                            'Вид работ':'work_type',
                            'Unnamed: 18':'Unk_col'
                           })
```

```python
# convert data type of column 'expert_date'
data['expert_date'] = pd.to_datetime(data['expert_date'])
```

```python
# check output
data.head(10)
```

```python
# function for searching ID numbers of companies across description
def id_number_selection(row):
    pattern = r"ИНН:\s\d{10}"
    sequence = str(row['designer_name'])
    try:
        result = re.findall(pattern,sequence)
        id_number = []
        for item in result:
            number = item[5:]
            id_number.append(number)
        if id_number == []:
            return np.nan
        else:
            return id_number
    except: return np.nan
```

```python
# applying function to data
data['designer_id_number'] = data.apply(id_number_selection,axis=1)
```

```python
# deleting rows with missing data in column 'designer_id_number'
data = data.dropna(axis='index', how='any', subset=['designer_id_number'])
```

```python
# exploding data in column designer_id_number to put each ID number for each row
data = data.explode('designer_id_number')
```

```python
# check top designers
top_designers = data['designer_id_number'].value_counts()
```

```python
# deleting unnecesary columns from dataset
data.drop(['doc_type', 'expert_name', 'builder_name', 'tech_cust_name', 'typical', 'typical2', 'typ_id', 'reestr_date','Unk_col'], axis=1, inplace=True)
```

```python
# reading data from external csv file with proper list of designers
designers = pd.read_csv('designers.csv', sep=';', dtype={'designer_name': object , 'designer_id_number': object})  
```

```python
# check reading data
designers.head(1)
```

```python
# merging table of raw data with table of proper designers
projects = designers.merge(data, on='designer_id_number', how='left')
```

```python
# deleting unnecesary columns from dataset
projects = projects.dropna(subset=['region_id','object'])
```

```python
# convert data type of column 'expert_date'
projects['region_id'] = projects['region_id'].astype('int')
```

```python
# preparing output file for dashboard
projects.to_csv('projects.csv', index=False, sep=';')
```
